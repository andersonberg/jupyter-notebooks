```python
import pandas as pd
import numpy as np

MAIN_ST_STATS = ('shoot_finishing', 'pace_acceleration', 'phys_strength', 'drib_dribbling')
MAIN_CB_STATS = ('pace_acceleration', 'phys_strength', 'def_stand_tackle', 'drib_agility')
MAIN_RLB_STATS = ('pace', 'drib_agility', 'pass_short', 'phys_strength')
MAIN_CM_STATS = ('phys_strength', 'pass_short', 'pace_acceleration', 'drib_dribbling')
MAIN_CDM_STATS = ('physicality', 'pace_acceleration', 'def_interceptions', 'def_stand_tackle')
MAIN_CAM_STATS = ('passing', 'drib_dribbling', 'shoot_long_shots', 'pace_acceleration')
MAIN_LRM_STATS = ('pace', 'phys_strength', 'pass_short', 'drib_dribbling')
MAIN_LRWF_STATS = ('pace', 'shoot_finishing', 'pass_short', 'drib_dribbling')
# MAIN_GK_STATS = ('gk_diving', 'gk_handling', 'gk_reflexes')
```

```python
complete_df = pd.read_csv('FIFA19 - Ultimate Team players.csv', encoding='utf_8')
list(complete_df.keys())
```

```python
all_players = complete_df.drop(complete_df.columns[[2, 5, 11, 12, 13, 14, 15, 16]], axis=1)
# all_players.head(200)
```

```python
def get_league_players(all_players, league_full_name):
    return all_players[(all_players['league']==league_full_name)]

def stats_mean(row, *stats):
    selected_stats = []
    for stat in stats:
        selected_stats.append(row[stat])
    return np.mean(selected_stats)

players_df = get_league_players(all_players, 'Premier League')
# players_df = all_players
```

```python
# strikers = players_df[(players_df['position']=='ST') & (players_df['quality']=='Gold - Rare') & (players_df['revision']=='Normal')]
strikers = players_df.query('position=="ST" or position=="CF"').query('overall>=79')
strikers = strikers.assign(mean=strikers.apply(stats_mean, axis=1, args=MAIN_ST_STATS))

sort_keys = list(MAIN_ST_STATS)
sort_keys.insert(0, 'mean')
display_keys = ['player_name', 'overall', 'revision']
display_keys.extend(sort_keys)
strikers.sort_values(sort_keys, ascending=False)[display_keys]
```

```python
# zagueiros = players_df[(players_df['position']=='CB') & (players_df['quality']=='Gold - Rare') & (players_df['revision']=='Normal')]
zagueiros = players_df.query('position=="CB"').query('overall>=79').query('revision=="Normal"')
zagueiros = zagueiros.assign(mean=zagueiros.apply(stats_mean, axis=1, args=MAIN_CB_STATS))

sort_keys = list(MAIN_CB_STATS)
sort_keys.insert(0, 'mean')
display_keys = ['player_name', 'overall']
display_keys.extend(sort_keys)
zagueiros.sort_values(sort_keys, ascending=False)[display_keys]
```
